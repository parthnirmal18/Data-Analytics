```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
x = np.array([12,23,34,45,49,56,61,68,71,78])

y = np.array([23,31,38,43,49,56,65,71,79,85])

z = np.array([98,89,86,76,71,68,61,54,49,34])
```

```python
plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.plot(x, y, marker = 'D', color = 'blue', mfc = 'r', ls = '--', ms = 7)
```

```python
plt.title('X vs Z')
plt.xlabel('First Value')
plt.ylabel('Third Value')
plt.plot(x, z, marker = 'p', color = 'maroon', mfc = 'yellow', ls = '-.', ms = 9)

plt.grid()
```

```python
plt.title('X vs Z')
plt.xlabel('First Value')
plt.ylabel('Second and Third Value')
plt.plot(x, z, marker = 'p', color = 'maroon', mfc = 'yellow', ls = '-.', ms = 9)
plt.plot(x, y, marker = 'D', color = 'blue', mfc = 'r', ls = '--', ms = 7)
plt.grid()
```

```python
plt.figure(figsize=(10,8))

plt.subplot(2,2,1)

plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.plot(x, y, marker = 'D', color = 'blue', mfc = 'r', ls = '--', ms = 7)


plt.subplot(2,2,2)

plt.title('X vs Z')
plt.xlabel('First Value')
plt.ylabel('Third Value')
plt.plot(x, z, marker = 'p', color = 'maroon', mfc = 'yellow', ls = '-.', ms = 9)

plt.grid()
```

```python
students = [56, 65, 46, 51, 61]

courses = ['Java', 'Python', 'R', 'Kotlin', 'Dart']
```

# Seaborn

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
import matplotlib as mpl
mpl.rcParams.update(mpl.rcParamsDefault)
```

```python
sns.jointplot(x = 'tip', y = 'total_bill', data = tips, hue = 'time')
plt.show()
```

```python
plt.xlabel('Tip Amount')
plt.ylabel('Frequency')
sns.histplot(tips['tip'],kde = True)
plt.show()
```

```python
sns.boxplot(tips['tip'])
plt.show()
```

```python
plt.title('Outliers in the Tip by Days')
plt.xlabel('Day')
plt.ylabel('Tip')
sns.boxenplot(x = 'day',y = 'tip', data = tips)
plt.show()
```

```python
sns.pairplot(tips , hue = 'smoker')

plt.show()
```
